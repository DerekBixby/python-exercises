```python
# 4. Creating Powers Table that asks if user wants to continue

def square(x):
    return x*x
def cube(x):
    return x*x*x

def powers_table():
    first = 1
    last = int(input('choose a number to go up to'))

    print('======\t\t=======\t\t======')
    print('Number\t\tSquare\t\tCube')
    print('======\t\t======\t\t======')

    for x in range(first, last+1):
        print(x,'\t\t', square(x), '\t\t', cube(x))

    while True:
        a = input('Do you want to continue?')
        if a =='yes':
            print('continue')
            continue
        elif a == 'no':
            break
        else:
            print('yes or no')
            
powers_table()

    
```

```python
# 5. Take in number grade and return letter grade. Prompt user on if they want to continue


while True:
    grade_prompt = int(input('Please enter a numerical grade from 0 to 100 '))

    if grade_prompt < 60:
        print('F')
    elif grade_prompt < 67 and grade_prompt > 59:
        print('D')
    elif grade_prompt < 80 and grade_prompt > 66:
        print('C')
    elif grade_prompt < 88 and grade_prompt > 79:
        print('B')
    elif grade_prompt > 87:
        print('A')
    input_continue = input('Do you want to continue?')

    if input_continue == 'n':
        break
    


```
